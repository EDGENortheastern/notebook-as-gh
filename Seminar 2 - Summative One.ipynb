## Synthetic Data Generation üß™

In this section, synthetic data is created to represent key operational measures from organic farms. [NumPy](https://numpy.org/doc/) is used because it allows controlled generation of numeric data that follows specific statistical patterns.

### Import libraries

```
import numpy as np # for data generation
import pandas as pd # for organising the data into a table
import matplotlib.pyplot as plt # for static data viz
import plotly.express as px # for interactive viz
```

### Set a random seed

A random seed is set to ensure the results can be reproduced. This means the same synthetic dataset can be generated again by re-running the notebook

```
np.random.seed(42)
rows = 120
```

### Generating the Year Data üìÖ

The farmers were recording the crops data between 2018 and 2925 therefore

```
years = np.arange(2018, 2026)
years
```

```
year = np.random.choice(years, size=rows)
year
```

```
summer_sunny_days = np.random.uniform(40, 90, rows)
```

```
soil_quality = np.random.uniform(1, 10, rows)
```

```
crop_output_tonnes = np.random.normal(
    loc=300 + (summer_sunny_days * 0.8),
    scale=40,
    size=rows
)
```

```
annual_revenue_gbp = np.random.normal(
    loc=crop_output_tonnes * 250,
    scale=15000,
    size=rows
)
```

```
farm_data = pd.DataFrame({
    "year": year,
    "summer_sunny_days": summer_sunny_days,
    "crop_output_tonnes": crop_output_tonnes,
    "annual_revenue_gbp": annual_revenue_gbp
})
farm_data
```

```
farm_data = farm_data.sort_values(by="year").reset_index(drop=True)
farm_data
```

```
yearly_summary = farm_data.groupby("year").mean(numeric_only=True)
```

```
plt.figure()
plt.plot(
    yearly_summary.index,
    yearly_summary["annual_revenue_gbp"],
    marker="o"
)
plt.title("Average Annual Revenue by Year")
plt.xlabel("Year")
plt.ylabel("Average Revenue (¬£)")
plt.show()
```

```
fig_hist = px.histogram(
    farm_data,
    x="summer_sunny_days",
    nbins=15,
    title="Distribution of Summer Sunny Days"
)

fig_hist.show()
```

```
fig_box = px.box(
    farm_data,
    y="crop_output_tonnes",
    title="Crop Output Distribution"
)

fig_box.show()
```
